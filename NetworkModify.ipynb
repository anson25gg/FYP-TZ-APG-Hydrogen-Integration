```python
p_set_values = {
    "IDN": 6500,
    "MYS": 5000,
    "PHL": 200,
    "SGP": 2000,
    "BRN": 150,
    "MMR": 100,
    "THA": 1200,
    "VNM": 2200,
    "KHM": 100,
    "LAO": 100
}

for country in ["IDN", "MYS", "PHL", "SGP", "BRN", "MMR", "THA", "VNM", "KHM", "LAO"]:
    n.add("Load",
      name=f"HydrogenDemand_{country}",
      p_set=p_set_values[country],
      bus=f"{country}_H2",
      carrier="Hydrogen")
```

```python
for bus in n.buses.index[0:24]:
    h2_bus = bus[:3] + "_H2"
    n.add("Link",
          name=f"Electrolyser-{bus}-to-{h2_bus}",
          bus0=bus,
          bus1=h2_bus,
          build_year=2023,
          lifetime=20,
          efficiency=0.7,
          p_nom=10,
          p_nom_extendable=True,
          capital_cost=120000.0,
          marginal_cost=71.43,
          carrier="Hydrogen")
    
```
